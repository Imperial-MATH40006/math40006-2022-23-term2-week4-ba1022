```python
from numpy import linspace
xvals = linspace(-1.2, 1.2, 100)
yvals = linspace(-1.2, 1.2, 100)
print(xvals)
```

```python
from numpy import meshgrid
x, y = meshgrid(xvals, yvals)
print(y)
```

```python
z = (y*(y - x**2 +1)*(y + x**2 -1))/(1+ x**2 + y**2)
print(z)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.axis('equal')
plt.contour(x, y, z)
```

```python
fig = plt.figure() 
ax = plt.axes(projection = '3d') 
ax.plot_wireframe(x, y, z)
```

```python
def gcd1(a, b):
    """
    Calculates the greatest common divisor of a and b
    using Euclid's algorithm.
    """
    
    # while loop
    while b > 0:
        a, b = b, a % b
        
    # return
    return a
```

```python
#a

print(gcd1(10,5))
print(gcd1(11,5))

#b
from random import randint
from time import time
factor = randint(10**19,10**20)
print(factor)
mult1 = randint(10**79,10**80)
mult2 = randint(10**79,10**80)
print(gcd1(mult1,mult2))
start = time()
print(gcd1(factor*mult1,factor*mult2))
print(time()-start)

#c two primes

print(gcd1(13,7))
```

```python
def gcd2(a, b):
    while True:
        if b==0:
            return a
        a, b = b, a % b        
```

```python
# typical cases
assert(gcd2(35,55)==5)
assert(gcd2(35,56)==7)
assert(gcd2(35,57)==1)
```
